## 16- Splitting a string into multiple columns

```python
import pandas as pd

df= pd.DataFrame({'name':['azka saleem','junaid latif','taeeda atvat','roulin yang'],
                    'location':['fsd,pk','kml,pk','kml,pk','xian,ch']})
                    
df
```

```python
df.name.str.split(' ',expand=True).head()
```

```python
df
```

```python
df[['first name','last name']]= df.name.str.split(' ',expand=True)
df
```

```python
df[['city','country']]= df.location.str.split(',',expand=True)
df
```

```python
df= df[['first name','last name','city','country']]
df
```

## 23- Reshaping a dataframe

```python
fasla= pd.DataFrame([['1234',100,200,300],['5678',400,500,600],['9012',700,800,900]],
                        columns=['zip','factory','warehouse','retail'])
fasla.head()
```

```python
fasla.T
```

```python
fasla2= pd.DataFrame([[1,'1234','factory'],[2,'5678','warehouse'],
                        [3,'9012','retail']],columns=['id','zip','location'])
fasla2.head()
```

```python
faslanew= fasla.melt(id_vars=['zip'],value_name='distance',var_name='location')
faslanew.head()
```

```python
import seaborn as sns
sns.barplot(x='zip',y='distance',hue='location', data=faslanew)
```
