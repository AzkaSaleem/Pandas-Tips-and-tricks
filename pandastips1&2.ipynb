## 03- How to rename columns

```python
df = bigdf.rename(columns={'A': 'a', 'B': 'b', 'C': 'c'})
df.head()
```

```python
df.columns= ['aa', 'bb', 'cc']
df.head()
```

```python
df.columns=df.columns.str.replace('a', 'A')
df.head()
```

```python
# add prefix in column names
df = df.add_prefix('col_')
df.head()
```

```python
# add suffix in column names
df = df.add_suffix('col_')
df.head()
```
